```python
import pandas as pd 

stores = pd.read_csv('../data/stores.csv')
features = pd.read_csv('../data/features.csv')
df = pd.read_csv('../data/train.csv')

stores.head()
features.head()
```

```python
pdf = pd.merge(df, features, on=['Store', 'Date', 'IsHoliday'], how='left')

pdf = pd.merge(pdf, stores, on=['Store'], how='left')

pdf.head()

pdf.to_csv('../data/train_merged.csv', index=False)
```
