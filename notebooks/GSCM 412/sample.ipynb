```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define parameters
num_salespeople = 5
num_days = 10

start_date = "2020-01-01"
dates = pd.date_range(start=start_date, periods=num_days, freq="D")

# Define cars and their prices
cars = {
    "Toyota Corolla": 22000,
    "Ford F-150": 55000,
    "Honda CRV": 29000,
    "Chevy Bolt": 27500,
    "Tesla Model S": 80000,
    "Nissan Rogue": 32000,
    "BMW 3 Series": 45000,
    "Audi Q5": 52000,
    "Mercedes-Benz C-Class": 48000,
    "Hyundai Elantra": 21000,
    "Kia Soul": 22000,
    "Subaru Outback": 27000,
    "Mazda CX-5": 28000,
    "Volkswagen Jetta": 25000,
    "Jeep Wrangler": 35000,
    "Ram 1500": 45000,
    "GMC Sierra": 48000,
    "Lexus RX": 52000,
    "Cadillac Escalade": 75000,
    "Porsche 911": 100000,
}

salespersons_data = {
    "Elite Motors Inc.": {
        "John": 0.05,
        "David": 0.08,
        "Olivia": 0.08,
        "Sophia": 0.05,
        "Charlie": 0.06,
    },
    "Prestige Auto Group": {
        "Ryan": 0.07,
        "Jonathan": 0.06,
        "Andrew": 0.06,
        "Nathan": 0.07,
        "Barb": 0.08,
    },
    "Luxe Cars Boutique": {
        "Susan": 0.08,
        "Lauren": 0.05,
        "Sam": 0.04,
        "Tyler": 0.06,
        "Jack": 0.08,
    },
    "Apex Motors": {
        "Joe": 0.07,
        "Trevor": 0.05,
        "Steve": 0.06,
        "Katie": 0.08,
        "Emily": 0.05,
    },
    "Grand Avenue Motors": {
        "Sarah": 0.06,
        "Parker": 0.08,
        "Tucker": 0.07,
        "Brooke": 0.08,
        "Grant": 0.07,
    },
    "Royal Auto Emporium": {
        "Caleb": 0.06,
        "Brent": 0.08,
        "Kaylee": 0.08,
        "Brooklyn": 0.04,
        "Sydney": 0.05,
    },
    "Summit Motors LLC": {
        "Chloe": 0.08,
        "Mya": 0.08,
        "Josh": 0.05,
        "Justin": 0.06,
        "Scott": 0.07,
    },
    "Premier Automotive Solutions": {
        "Grace": 0.06,
        "Vincent": 0.07,
        "Megan": 0.08,
        "Spencer": 0.07,
        "Greg": 0.05,
    },
    "Regal Motors Corporation": {
        "Taylor": 0.05,
        "Mark": 0.06,
        "Ezra": 0.07,
        "Kayla": 0.08,
        "James": 0.05,
    },
    "Capital City Cars Dealership": {
        "Kim": 0.08,
        "Warren": 0.07,
        "Brandon": 0.08,
        "Nick": 0.07,
        "Nate": 0.05,
    },
}


weekday_adjustment = {
    "Monday": 0.9,
    "Tuesday": 0.95,
    "Wednesday": 1.0,
    "Thursday": 1.0,
    "Friday": 1.1,
    "Saturday": 1.3,
    "Sunday": 1.2,
}


# Generate random mean sales and standard deviation for each salesperson
mean_sales = np.random.uniform(6000, 12000, num_salespeople)
std_dev_sales = np.random.uniform(500, 1500, num_salespeople)

# Monte Carlo simulation parameters
num_trials = 100  # Number of trials

# Noise parameters
noise_mean = 2500
noise_std_dev = 1000

# Initialize list to store sales data across all trials
all_sales_data = []

for _ in range(num_trials):
    # Generate random sales data with car type
    sales_data = []
    for day in dates:
        # Random number of sales per day (between 5 and 25)
        for dealership in salespersons_data:
            sales_per_day = round(
                np.random.randint(5, 25) * weekday_adjustment[day.day_name()]
            )
            for iCount in range(sales_per_day):
                sale = {}
                car = np.random.choice(
                    list(cars.keys())
                )  # Randomly choose a car from the list
                sale["salesperson"] = np.random.choice(
                    list(salespersons_data[dealership].keys())
                )
                sale["car"] = car
                sale["price"] = cars[car]
                sale["date"] = day
                sale["weekday"] = day.day_name()
                sale["commission_rate"] = salespersons_data[dealership][
                    sale["salesperson"]
                ]
                sale["commission"] = sale["price"] * sale["commission_rate"]
                # Add noise to sales data
                sale["price"] += np.random.normal(noise_mean, noise_std_dev)
                sales_data.append(sale)
    # Append sales data of each trial to the list
    all_sales_data.extend(sales_data)

# Create DataFrame for all sales data across all trials
df_sales = pd.DataFrame(all_sales_data)

# Plot sales data for each salesperson
# for salesperson in salespersons_data:
#     sales_data_person = df_sales[df_sales["salesperson"] == salesperson]
#     plt.plot(sales_data_person.groupby("date").size(), label=salesperson)

# plt.title("Sales Over Time")
# plt.xlabel("Date")
# plt.ylabel("Number of Sales")
# plt.legend()
# plt.grid(True)
# plt.xticks(rotation=45)
# plt.tight_layout()
# plt.show()

# Display DataFrame
df_sales.head()
```
