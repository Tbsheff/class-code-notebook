```python
import requests


def get_coordinates_osm(city):
    params = {"q": city, "format": "json"}
    response = requests.get("https://nominatim.openstreetmap.org/search", params=params)
    if response.status_code == 200:
        data = response.json()
        if data:
            return city, float(data[0]["lat"]), float(data[0]["lon"])
    return city, None, None


cities = [
    "New York, NY",
    "Los Angeles, CA",
    "Denver, CO",
    "Minneapolis, MN",
    "Portland, OR",
    "Austin, TX",
    "Charlotte, NC",
]

medium_cities_us = [
    "Spokane, WA",
    "Boise, ID",
    "Reno, NV",
    "Provo, UT",
    "Albuquerque, NM",
    "Tucson, AZ",
    "Colorado Springs, CO",
    "Lubbock, TX",
    "Wichita, KS",
    "Omaha, NE",
    "Des Moines, IA",
    "Madison, WI",
    "Grand Rapids, MI",
    "Akron, OH",
    "Toledo, OH",
    "Rochester, NY",
    "Syracuse, NY",
    "Worcester, MA",
    "Richmond, VA",
    "Greensboro, NC",
    "Durham, NC",
    "Charleston, SC",
    "Jacksonville, FL",
    "Baton Rouge, LA",
    "Little Rock, AR",
    "Tulsa, OK",
    "Fayetteville, AR",
    "Springfield, MO",
    "Shreveport, LA",
    "Mobile, AL",
    "Knoxville, TN",
    "Chattanooga, TN",
    "Lexington, KY",
    "Huntsville, AL",
    "Columbia, SC",
    "Savannah, GA",
    "Tallahassee, FL",
    "Montgomery, AL",
    "Birmingham, AL",
    "Asheville, NC",
]


city_coordinates_osm = [get_coordinates_osm(city) for city in cities]
medium_cities_us_coordinates_osm = [get_coordinates_osm(city) for city in medium_cities_us]
print(city_coordinates_osm)
```

```python
import folium

m = folium.Map(location=[39.8283, -98.5795], tiles="Cartodb Positron", zoom_start=4)
m2 = folium.Map(location=[39.8283, -98.5795], tiles="Cartodb Positron", zoom_start=4)  # medium cities

for city, lat, lon in city_coordinates_osm:
    # folium.Marker([lat, lon], tooltip=city).add_to(m)
    # Plotting a circle around each city
    folium.Circle(
        location=[lat, lon],
        radius=700000,  # radius in meters (50 km)
        color="blue",
        fill=True,
        fill_color="blue",
        weight=0,
    ).add_to(m)

for city, lat, lon in medium_cities_us_coordinates_osm:
    # folium.Marker([lat, lon], tooltip=city).add_to(m2)
    folium.Circle(
        location=[lat, lon],
        radius=150000,  # radius in meters (50 km)
        color="red",
        fill=True,
        fill_color="red",
        weight=0,
    ).add_to(m2)



# Save to an HTML file
m.save("us_cities_map.html")
m2.save("us_medium_cities_map.html")
```
