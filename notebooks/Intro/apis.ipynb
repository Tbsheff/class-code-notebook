```python
import requests, json
import pandas as pd
```

```python
url = "https://baseball4.p.rapidapi.com/v1/mlb/games-playbyplay"

querystring = {"gamePk": "633282"}

headers = {
    "X-RapidAPI-Key": "bde58bb7b6msh799b2072251aa72p1504f4jsnb38205b8ffcb",
    "X-RapidAPI-Host": "baseball4.p.rapidapi.com",
}
response = requests.get(url, headers=headers, params=querystring)

json_response = json.loads(response.text)
print(json_response)

json_response = json_response["body"]

sorted_data = []
for i in range(len(json_response)):
    row= {}
    row["resultType"] = json_response[i]["result"]["type"]
    row["resultEvent"] = json_response[i]["result"]["event"]
    row["resultEventType"] = json_response[i]["result"]["eventType"]
    row["resultDescription"] = json_response[i]["result"]["description"]
    row["resultRbi"] = json_response[i]["result"]["rbi"]
    row["awayScore"] = json_response[i]["result"]["awayScore"]
    row["homeScore"] = json_response[i]["result"]["homeScore"]
    row["isScoringPlay"] = json_response[i]["about"]["isScoringPlay"]
    row["atBatIndex"] = json_response[i]["about"]["atBatIndex"]
    row["halfInning"] = json_response[i]["about"]["halfInning"]
    row["inning"] = json_response[i]["about"]["inning"]
    row["startTime"] = json_response[i]["about"]["startTime"]
    row["endTime"] = json_response[i]["about"]["endTime"]
    row["balls"] = json_response[i]["count"]["balls"]
    row["strikes"] = json_response[i]["count"]["strikes"]
    row["outs"] = json_response[i]["count"]["outs"]
    row["batterId"] = json_response[i]["matchup"]["batter"]["id"]
    row["batterFullName"] = json_response[i]["matchup"]["batter"]["fullName"]
    row["batSide"] = json_response[i]["matchup"]["batSide"]["description"]
    row["pitcherId"] = json_response[i]["matchup"]["pitcher"]["id"]
    row["pitcherFullName"] = json_response[i]["matchup"]["pitcher"]["fullName"]
    row["pitchHand"] = json_response[i]["matchup"]["pitchHand"]["description"]
    sorted_data.append(row)
```

```python
df = pd.DataFrame(sorted_data)
df
```

```python
df.to_csv("../outputs/playbyplay.csv", index=False)
```
