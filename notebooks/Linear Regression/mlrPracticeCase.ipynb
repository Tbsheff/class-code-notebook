```python
import pandas as pd
from sklearn.linear_model import (
    LinearRegression,
    Ridge,
    Lasso,
    LassoLars,
    BayesianRidge,
    TweedieRegressor,
)

from helperFunctions import regression


df_disney = pd.read_csv("../data/disney_movies_clean.csv")
df_disney.head()
```

```python
df_dummy = pd.get_dummies(
    df_disney.drop(columns=["movie_title", "release_date"]),
    drop_first=True,
    dtype="int64",
)
df_dummy
```

```python
regression(df_dummy.drop(columns=['total_gross']), "inflation_adjusted_gross")
```

```python
y = df_dummy["inflation_adjusted_gross"]
X = df_dummy.drop(columns=["inflation_adjusted_gross", "total_gross"])

model = TweedieRegressor(power=1, link="log", max_iter=1000).fit(X, y)
score = model.score(X, y)
print(score)
print(model.coef_)

```
